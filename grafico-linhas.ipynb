```python
#importação de bibliotecas
import random
import matplotlib.pyplot as plt
#gerar uma lista de 10 numeros aleatorios de 1 a 100
dados = [random.randint(1, 100) for _ in range (10)]

#exibir a lista gerada
print("lista de dados:", dados)

#plotar os dados
plt.plot(dados, marker='o') #marcador em cada ponto para melhorar visualização
plt.title("grafico de numeros aleatorios")
plt.xlabel("indice")
plt.ylabel("valor")
plt.grid(True)
plt.show()
```
